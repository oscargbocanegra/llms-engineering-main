```python
import sqlite3
from pathlib import Path

db_file = Path("/giova/llms-engineering-main/vector_db/chroma.sqlite3")
print("DB exists:", db_file.exists(), "size:", db_file.stat().st_size)

con = sqlite3.connect(str(db_file))
cur = con.cursor()
cur.execute("SELECT name FROM sqlite_master WHERE type='table' ORDER BY name;")
tables = [r[0] for r in cur.fetchall()]
con.close()

print("Tables:", tables)
print("Has collections table?", "collections" in tables)
```
